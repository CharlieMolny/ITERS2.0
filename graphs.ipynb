```python
import pandas as pd
import matplotlib.pyplot as plt



# Step 1: Read the CSV file, using the first column as the DataFrame index
df = pd.read_csv(r'eval\highway\best_summary_expl\IRS.csv', index_col=0)

print(df.head())
```

```python


# Now, the DataFrame should be correctly aligned, proceed with plotting
plt.figure(figsize=(10, 6))

# Plotting each lambda value on the same figure
for lmbda in df['lmbda'].unique():
    df_lmbda = df[df['lmbda'] == lmbda]
    plt.plot(df_lmbda['iter'], df_lmbda['true_reward'], label=f'={lmbda}')

plt.xlabel('Iteration')
plt.ylabel('True Reward')
plt.title('True Reward vs. Iteration for each ')
plt.legend()
plt.show()
```
